```python
mean_chl = data.Cholesterol.mean()
mean_chl
```

```python
std_chl = data.Cholesterol.std()
std_chl
```

```python
data['zscore'] = (data.Cholesterol-mean_chl)/std_chl
data.head()
```

```python
data.zscore.max()
```

```python
data.zscore.min()
```

```python
filtered_data = data[(data.zscore<3)&(data.zscore>-3)]
filtered_data
```

```python
upper_limit = mean_chl + 3*std_chl
lower_limit = mean_chl - 3*std_chl
upper_limit,lower_limit
```

```python
fdata = data[(data.Cholesterol < upper_limit) & (data.Cholesterol>lower_limit)]
fdata
```

```python
y = ['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope']
from sklearn.preprocessing import LabelEncoder
lr = LabelEncoder()
for x in y:
    data[x,'_enc'] = lr.fit_transform(data[x])
data.head()
```

```python
f_data  = data.drop(data[['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope']], axis = 'columns')
f_data.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
y=f_data.drop(['HeartDisease'], axis='columns')
y.columns =y.columns.astype(str)
```

```python
scaled_data=ss.fit_transform(y)
f_scale_data = pd.DataFrame(scaled_data, columns = y.columns)
f_scale_data.head()
```

```python
final_data = f_scale_data
target = data['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(final_data, target,test_size=0.25, random_state=10)
```

```python
len(x_train), len(x_test)
```

```python
from sklearn.svm import SVC
svm =SVC()
svm.fit(x_train, y_train)
svm.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(SVC(), x_train, y_train, cv=5)
score
```

```python
score.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier(
        estimator =SVC(),
        n_estimators=100,
        max_samples=0.8,
        oob_score=True,
        random_state=0
)
bag_model.fit(x_train, y_train)
bag_model.score(x_test, y_test)
```

As you can see there is no as such differece in SVM and bagged score.Now lets try Decision Tree Classifier.

```python
from sklearn.tree import DecisionTreeClassifier
dtc= DecisionTreeClassifier(max_depth =3)
dtc.fit(x_train, y_train)
dtc.score(x_test, y_test)
```

```python
score_dtc = cross_val_score(DecisionTreeClassifier(random_state=0), x_train, y_train, cv=5)
score
```

```python
score.mean()
```

```python
bag_model_dtc = BaggingClassifier(
        estimator =DecisionTreeClassifier(),
        n_estimators=100,
        max_samples=0.8,
        oob_score=True,
        random_state=0
)
bag_model_dtc.fit(x_train, y_train)
bag_model_dtc.score(x_test, y_test)
```
